```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import glob
```

```python
# Define tile IDs, models being compared

tile_id = ['E-20', 'B-19', 'D-17', 'D-15', 'G-15']
ml_models = ['gm_mads_two_seasons', 'gm_mads_xy_two_seasons']
region = ['Eastern']
```

```python
# Load TIF files from Classifications folder
# Uses approx 50gb of RAM, use unstable sandbox

fig, ax = plt.subplots(len(tile_id), len(ml_models), figsize=(26, 60))

for i in range(len(tile_id)):
    for j in range(len(ml_models)):
        for k in range(len(region)):
            filename = region[k]+'_tile_'+tile_id[i] +'_prediction_pixel_' + ml_models[j] + '*.tif'
            filename = glob.glob(str(filename))
            x = xr.open_rasterio(filename[0]).squeeze()
            x.plot(ax=ax[i,j], add_colorbar=False)
            ax[i,j].set_title(region[k]+tile_id[i]+ml_models[j], size = 8)
            
```
